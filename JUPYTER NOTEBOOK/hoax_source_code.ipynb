# Pembelajaran Mesin

## Data Split - Hold-out Validation

```python
from sklearn.model_selection import train_test_split
```

```python
X = features_df.iloc[:200] # mengambil atribut/fitur data
Y = indexed_df.iloc[:200:,-1]#mengambil target data
```

```python
X
```

```python
pd.DataFrame(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.3, random_state=5)
```

```python
print("Data latih : ",X_train.shape)
print("Data Uji : ",X_test.shape)
```

```python
X_train
```

```python
X_test
```

```python
pd.DataFrame(Y_train)
```

```python
pd.DataFrame(Y_test)
```

## Algoritma Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
model_gnb = gnb.fit(X_train,Y_train)
```

```python
pred_gnb = model_gnb.predict(X_test)
```

```python
X_test
```

```python
print("Prediksi Algoritma Naive Bayes")
pd.DataFrame(data=pred_gnb,index=X_test.index,columns=[indexed_df.columns.values[-1]])
```

```python
print("Label Sebenarnya :")
pd.DataFrame(data=Y_test,index=X_test.index)
```

```python
pred_prob_gnb = model_gnb.predict_proba(X_test)
```

```python
pd.DataFrame(data=pred_prob_gnb,index=X_test.index,columns=["Hoaks","Non-Hoaks"])
```

# Algoritma Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb = MultinomialNB()
model_mnb = mnb.fit(X_train, Y_train)
```

```python
pred_mnb = model_mnb.predict(X_test)
```

```python
X_test
```

```python
print("Prediksi Algoritma Multinomial Naive Bayes")
pd.DataFrame(data=pred_mnb,index=X_test.index, columns=[indexed_df.columns[-1]])
```

```python
print("Label Sebenarnya")
pd.DataFrame(data=Y_test,index=X_test.index)
```

```python
pred_prob_mnb = model_mnb.predict_proba(X_test)
pd.DataFrame(data=pred_prob_mnb,index=X_test.index,columns=['Hoaks','Non-Hoaks'])
```

# Evaluasi

## Naive Bayes

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_cm_gnb =  pd.crosstab(Y_test, pred_gnb, rownames=['Actual'], colnames=['Predicted'])
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_cm_gnb, annot=True,annot_kws={"size": 16})# font size
```

```python
print(classification_report(Y_test, pred_gnb))
```

```python
print("Nilai Akurasi Gaussian Naive Bayes: ", "%.2f"% accuracy_score(Y_test,pred_gnb))
```

```python
from sklearn.metrics import f1_score
print("Nilai F1 Gaussian Naive Bayes: ", "%.2f"% f1_score(Y_test,pred_gnb))
```

## Multinomial Naive Bayes

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_cm_mnb = pd.crosstab(Y_test, pred_mnb, rownames=['Actual'], colnames=['Predicted'])
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_cm_mnb, annot=True,annot_kws={"size": 16})# font size
```

```python
print(classification_report(Y_test, pred_mnb))
```

```python
print("Nilai Akurasi Multinomial Naive Bayes: ", "%.2f"% accuracy_score(Y_test,pred_mnb))
```

```python
from sklearn.metrics import f1_score
print("Nilai F1 Multinomial Naive Bayes: ", "%.2f"% f1_score(Y_test,pred_mnb))
```
